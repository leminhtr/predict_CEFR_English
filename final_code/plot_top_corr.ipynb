```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
top_corr = pd.read_csv('./rapport/top_corr.csv', header=None)
top_corr.columns = ['Variable 1', 'Variable 2', 'Corrélation']
```

```python
top_corr.head()
top_corr.shape
```

```python
corr_value_list = top_corr['Corrélation']
```

```python
fig, (ax1) = plt.subplots(1,  figsize=(15,7))
fig.suptitle('Répartition des corrélations des couples de variables (décroissant)')

# Explained variance desc sort
ax1.plot(corr_value_list)

# Add horizontal line for y
# plt.axhline(y=1.0, linewidth=1, color='black')
plt.axhline(y=0.90, linewidth=0.5, color='r')
plt.axhline(y=0.75, linewidth=0.5, color='r')
plt.axhline(y=0.5, linewidth=0.5, color='r')
plt.axhline(y=0.25, linewidth=0.5, color='r')

# Tick frequency every 0.1 for y axis
plt.yticks(np.arange(min(corr_value_list), max(corr_value_list)+1, 0.1))

ax1.plot()
ax1.set_xlabel('Couple de variables')
ax1.set_ylabel('Corrélation')
ax1.set_ylim(ymin=0, ymax=1)
ax1.set_xlim(xmin=0)

fig.subplots_adjust(hspace=0)
fig.savefig('./rapport/top_corr_plot.png')
```
